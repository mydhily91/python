```python
# INTRODUCTION TO PANDAS
# SERIES
'''
Pandas is a python libaries that provides fast and easy data anayalsis platform
'''
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
# A series is like an array in Numpi except it has data labels so it's indexed.
obj = Series([3,6,9,12])
print obj
print("\n if u want to only the values we can call obj.values \n")
print obj.values
print "\n If you want to see the index we can call obj.index \n"
print obj.index
# now we are going to create a series for world war 11 casualities
ww2 = Series([8700000,4300000,300000,2100000,400000],index = ['USSR','Germany','China','Japan','USA'])
print "\n THIS HOW THE SERIES LOOK LIKE \n"
print ww2
print "\n IF YOU WANT TO WHICH COUNTRY HAS SUFFERED CASULITIES OF THIS SERIES \n"
print ww2['USSR']
print "\n Check which casualites are greater than 4 million \n"
print ww2[ww2 > 400000]
# We can also treat a series as an ordered dictionary in Python.
# So for instance if we wanted to just check if a index or value was in a series or not
print "\n check if a index or value was in a series or not \n"
print 'USSR' in ww2
# Since it kind of behaves a bit like a dictionary, we can actually convert a series into a dictionary.
print "\n To convert series into dictionary \n"
ww2_dict = ww2.to_dict()
print ww2_dict
print "\n To convert back into the series \n"
ww2_series = Series(ww2_dict)
print ww2_series
# If you were gonna pass a dictionary into the series. The index will have the dictionary keys in order.
print "\n Create a new list \n"
Countries = ['China','Germany','India','USSR','USA','SriLanka']
obj2 = Series(ww2_dict,index = Countries)
print obj2
# Pandas has built-in methods to find null and replace them or check through there or not.
# Those are is null and not null. we can call pd.isnull
print "\n To find the Null \n"
print pd.isnull(obj2)
print "\n To opp of the above wheather it not null \n"
print pd.notnull(obj2)
#Another thing we can do in Pandas,is we can add Series to Series,
# and it'll automatically align the data by index.
print "\n Adding up two series \n"
print ww2_series
print "\n"
print ww2_series + obj2
# To add name to object 
print "\n Adding name to the object2 \n"
obj2.name = 'World War 2 Casulaties'
print obj2
print "\n We call also name the index \n"
obj2.index.name = 'Countries'
print obj2
```
